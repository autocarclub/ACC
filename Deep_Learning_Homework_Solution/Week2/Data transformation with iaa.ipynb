```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
import glob
import cv2

import imgaug.augmenters as iaa

import matplotlib.pyplot as plt
print(os.listdir("../input"))

# Any results you write to the current directory are saved as output.
```

```python
train_folder = "../input/catndog/catndog/train/"
test_folder = "../input/catndog/catndog/test"
```

```python
classes = ["cat", "dog"]
resized_width = resized_height = 32
```

```python
def preprocessing(img):
    return cv2.resize(img, (resized_width, resized_height))

def load_data_label_nparray(folder):
    list_img = []
    list_label = []
    for i, cls in enumerate(classes):
        fnames = glob.glob(os.path.join(folder, cls, "*.jpg"))
        for fname in fnames:
            img = plt.imread(fname)
            img = preprocessing(img)
            list_img.append(img)
            list_label.append(i)
    
    nparray_images = np.array(list_img)
    nparray_labels = np.array(list_label)
    
    return nparray_images, nparray_labels
```

```python
x_train, y_train = load_data_label_nparray(train_folder)
x_test, y_test = load_data_label_nparray(test_folder)
```

```python
x_train.shape
```

```python
seq = iaa.Sequential([
    iaa.Crop(px=(0, 16)), # crop images from each side by 0 to 16px (randomly chosen)
    iaa.Fliplr(0.5), # horizontally flip 50% of the images
    iaa.GaussianBlur(sigma=(0, 3.0)) # blur images with a sigma of 0 to 3.0
])
```

```python
x_aug = seq.augment_images(x_train)
```
