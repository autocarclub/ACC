# b. Testing phase:
- We take numbers of random images in test set and predict their labels, and imshow them
- You don't have to write code this phase, I have done it for you, just enjoy

```python
# testing phase
test_folder = "./data/test/"

classes = ["left", "right", "stop"]

data = []
label = []

img_W = 32
img_H = 32

# number of image you wanna visualize and test
n_test = 100

def preprocessing(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return cv2.resize(gray, (img_W, img_H))

for i in range(len(classes)):
    class_name = classes[i]
    print("Now loading class: ", class_name)
    file_names = glob.glob(test_folder+class_name+"/*.ppm")
    for file_name in file_names:
        img = cv2.imread(file_name)
        img = preprocessing(img)
        data.append(img)
        label.append(i)
        #cv2.imshow("img", img)
        #cv2.waitKey(0)
        
#convert data and label to numpy array
X = np.array(data).astype(np.float32)
y = np.array(label).astype(np.int64)
        
# reshape data to (num_of_imgs, img_W * img_H)
X = X.reshape(-1, img_W*img_H)

n_images = X.shape[0]
print(n_images)

if(n_test > n_images):
    print("Number of sample image must <= total number of test image: ",n_images)
else:
    indices = np.random.permutation(n_images)[:n_test]
    for idx in indices:
        x = X[idx,:].reshape(1,-1)
        img = data[idx]
        y_true = y[idx]
        y_pred = svm_model.predict(x)
        print("===================================================")
        print("Your prediction label: ", classes[int(y_pred)],"\nActual label: ", classes[int(y_true)])
        cv2.imshow("img", img)
        cv2.waitKey(0)
```
